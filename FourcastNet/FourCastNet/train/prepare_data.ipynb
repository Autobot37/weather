```python
import h5py
import numpy as np
```

```python
file_path = '/home/vatsal/Supreme/Fourcastnet_outofsample/out_of_sample/2018.h5'
with h5py.File(file_path, 'r') as f:
    data = f['fields'][:]
```

```python
latitudes = np.load("/home/vatsal/NWM/FourcastNet/stats_v0/latitude.npy")
```

```python
longitudes = np.load("/home/vatsal/NWM/FourcastNet/stats_v0/longitude.npy")
```

```python
lat_min, lat_max = 6.37, 10.71
lon_min, lon_max = 74.68674, 79.023
```

```python
lat_indices = np.where((latitudes >= lat_min) & (latitudes <= lat_max))[0]
lon_indices = np.where((longitudes >= lon_min) & (longitudes <= lon_max))[0]
```

```python
print("Latitude indices:", lat_indices)
print("Longitude indices:", lon_indices)
```

```python
lat_start, lat_end = lat_indices[0], lat_indices[-2] + 1
lon_start, lon_end = lon_indices[0], lon_indices[-3] + 1
```

```python
sliced_data = data[..., lat_start:lat_end, lon_start:lon_end]
```

```python
num_data_points = sliced_data.shape[0]
train_data = sliced_data[:int(num_data_points * 0.9)]
val_data = sliced_data[int(num_data_points * 0.9):]
```

```python
with h5py.File("finetunetrainfile", 'w') as f_out:
    f_out.create_dataset('fields', data=train_data)
```

```python
with h5py.File("finetunevalfile", 'w') as f_out:
    f_out.create_dataset('fields', data=val_data)
```

```python
with h5py.File("finetunetrainfile", 'r') as f_in:
    val_data = f_in['fields'][:]
    print("Validation data shape:", val_data.shape)
    print("Validation data:", val_data)
```
